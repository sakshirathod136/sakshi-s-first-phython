```python
# Program to demonstrate type casting

a = 10        # integer
b = 5.5       # float

print("Original value of a:", a, "Type:", type(a))
print("Original value of b:", b, "Type:", type(b))

# Type casting
a = float(a)
b = int(b)

print("After type casting a:", a, "Type:", type(a))
print("After type casting b:", b, "Type:", type(b))
```
