```python
# Program to find greatest of 3 numbers

a = 10
b = 25
c = 18

print("Value of a:", a)
print("Value of b:", b)
print("Value of c:", c)

if a > b and a > c:
    print("Greatest number is:", a)
elif b > a and b > c:
    print("Greatest number is:", b)
else:
    print("Greatest number is:", c)
```
